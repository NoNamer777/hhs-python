```python
#Creating lists from columns in dataframe
import pandas as pd
import numpy as np
storedata = pd.read_csv('storedata_clean.csv', encoding = "ISO-8859-1",sep=";")

turnover = storedata['Profit'].values.tolist()
staff = storedata['Staff'].values.tolist()
floor = storedata['Floor Space'].values.tolist()
comp_sc = storedata['Competition score'].values.tolist()
min30 = storedata['30 min population'].values.tolist()
```

```python
#import library
import matplotlib.pyplot as plt
```

```python
#scatter staff/turnover
staff_turnover_plot = plt.scatter(staff, turnover)

plt.show()
```

```python
#scatter floor/turnover
floor_turnover_plot = plt.scatter(floor, turnover)

plt.show()
```

```python
#scatter comp_sc/turnover
comp_sc_turnover_plot = plt.scatter(comp_sc, turnover)

plt.show()
```

```python
# pie diagram
# Print the Staff array
print("\nThe number of employees per store:")
print(staff)

# Calculate the max of the Staff array.
staff_arr = np.array(staff)
max_staff = np.amax(staff_arr)

print("\nThe maximum number of employees a store can have:")
print(max_staff)

# Create a list to count how often the numbers occur in the staff List.
# Use the list.count function
max_staff_occ = np.count_nonzero(staff_arr == max_staff)

print("\nHow many stores have the same number of employees as the maximum number of employees a store can have:")
print(max_staff_occ)

# Create a List y for the labels of the pie
dist_employees_per_store = np.unique(staff_arr, return_index=False, return_inverse=False, return_counts=False)

print("\nThe number of employees a store has, according to the data:")
print(dist_employees_per_store)

# Create a List x to draw the slices within the pie
occ_employees_per_store = np.unique(staff_arr, return_index=False, return_inverse=False, return_counts=True)[1]

print("\nHow many times it occurs a store how a certain number of employees:")
print(occ_employees_per_store)

# Explode only second slice
exploded_slices = []

for idx, entry in enumerate(dist_employees_per_store):
    exploded_slices.append(0.2 if idx == 1 else 0)

# Define color scheme for slices of the pie diagram
employee_per_store_pie_colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k']

# plot the pie diagram
plt.pie(
    occ_employees_per_store,
    colors=employee_per_store_pie_colors,
    explode=exploded_slices,
    labels=dist_employees_per_store,
    startangle=90
)

plt.legend(title="Employees per store")

plt.show()
```
